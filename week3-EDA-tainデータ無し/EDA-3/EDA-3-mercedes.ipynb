# バイナリ変数を調べる

```python
# 空のリスト作成
zero_count_list = []
one_count_list = []
cols_list = unique_values_dict['[0, 1]']
for col in cols_list:
    zero_count_list.append((train_df[col]==0).sum())
    one_count_list.append((train_df[col]==1).sum())

N = len(cols_list)
ind = np.arange(N)
# 幅
width = 0.35

plt.figure(figsize=(6,100))

#plt.bashで横向きの棒グラフ
#indに縦列の位置zero_countに横の大きさ
#plt.legendは凡例表示するメゾット(p1,p2)
p1 = plt.barh(ind, zero_count_list, width, color='red')
p2 = plt.barh(ind, one_count_list, width, left=zero_count_list, color="blue")
plt.yticks(ind, cols_list)
plt.legend((p1[0], p2[0]), ('Zero count', 'One Count'))
plt.show()
```

さて、各バイナリ変数の平均値を調べてみましょう

```python
zero_mean_list = []
one_mean_list = []
cols_list = unique_values_dict['[0, 1]']
for col in cols_list:
    #meanで平均算出
    zero_mean_list.append(train_df.ix[train_df[col]==0].y.mean())
    one_mean_list.append(train_df.ix[train_df[col]==1].y.mean())

#Dataframeの作成
new_df = pd.DataFrame({"column_name":cols_list+cols_list, "value":[0]*len(cols_list) + [1]*len(cols_list), "y_mean":zero_mean_list+one_mean_list})

#ピボットテーブルの作成
new_df = new_df.pivot('column_name', 'value', 'y_mean')

plt.figure(figsize=(8,80))
sns.heatmap(new_df)
plt.title("Mean of y value across binary variables", fontsize=15)
plt.show()
```

上記のグラフで0と1の間に良好な色差を示すバイナリ変数は、両方のクラス間でカウント分布が良好である（前のグラフから見ることができる）ので、より予測的である可能性が高い。私たちは、ノートブックの後半部分で重要な変数について詳しく調べていきます。

**重要な変数：**

今私たちはxgboostモデルを実行して重要な変数を取得しましょう。

```python
for f in ["X0", "X1", "X2", "X3", "X4", "X5", "X6", "X8"]:
        #preprocessing.LabelEncoder()は文字データを離散の数値に変換する
        #またカテゴリカルデータを、数値(ラベル)で表現するメソッド。
        lbl = preprocessing.LabelEncoder()
        
        #sklearn.fit（）　fit関数で学習を行う
        lbl.fit(list(train_df[f].values)) 
        
        #sklearn.transformはパラメータをもとにデータ変換するメゾット
        train_df[f] = lbl.transform(list(train_df[f].values))
        
train_y = train_df['y'].values
#train_X指定の列を削除
train_X = train_df.drop(["ID", "y", "eval_set"], axis=1)

# Thanks to anokas for this #
def xgb_r2_score(preds, dtrain):
    #get_label（）DMatrixのラベルを取得します。
    labels = dtrain.get_label()
    return 'r2', r2_score(labels, preds)

#パラメーターの設定
xgb_params = {
    'eta': 0.05,#学習率を調整できる。
    'max_depth': 6,#木の深さの最大値
    'subsample': 0.7,#各木においてランダムに抽出される標本の割合小さくすることで、過学習を避けることができるが保守的なモデルとなる。
    'colsample_bytree': 0.7,#各木においてランダムに抽出される列の割合
    'objective': 'reg:linear',#最小化させるべき損失関数を指定する。
    'silent': 0 #モデルの実行結果を出力するかどうかを決めることができる。
}

#dtrain（DMatrix） - トレーニングDMatrixです。
dtrain = xgb.DMatrix(train_X, train_y, feature_names=train_X.columns.values)
#num_boost_roundは試行回数？フィットするブーストされたツリーの数。
# feval = カスタム評価関数。
#maximize = fevalを最大にするか。
model = xgb.train(dict(xgb_params, silent=0), dtrain, num_boost_round=100, feval=xgb_r2_score, maximize=True)

# plot the important features #
fig, ax = plt.subplots(figsize=(12,18))

#score をプロット
xgb.plot_importance(model, max_num_features=50, height=0.8, ax=ax)
plt.show()
```

カテゴリは、2進変数の後に続くトップスポットを占有します。

ランダムフォレストモデルを構築し、重要な変数をチェックしましょう。

```python
#ensembleはアンサンブルメゾット
from sklearn import ensemble

#ランダムフォレストとはアンサンブル学習法
#ランダムフォレストでは、アンサンブルの各ツリーは、トレーニングセットの置換（つまり、ブートストラップサンプル）で描かれたサンプルから作成されます。
#n_estimators　= 木の数
#max_depth = ツリーの最大深度
#min_samples_leaf = 葉ノードに必要なサンプルの最小数
#max_features = 最適な分割を探す際に考慮する機能の数
#n_jobs フィットと予測の両方を同時に実行するジョブの数。
#random_state = 乱数ジェネレータによって使用されるシード
model = ensemble.RandomForestRegressor(n_estimators=200, max_depth=10, min_samples_leaf=4, max_features=0.2, n_jobs=-1, random_state=0)

model.fit(train_X, train_y)
feat_names = train_X.columns.values

## 輸入量をプロットする ##
importances = model.feature_importances_

#np.stdは要素の標準偏差を求めるメゾット
std = np.std([tree.feature_importances_ for tree in model.estimators_], axis=0)

#argsortはソートした結果とインデックスを返す
indices = np.argsort(importances)[::-1][:20]

plt.figure(figsize=(12,12))
plt.title("Feature importances")
plt.bar(range(len(indices)), importances[indices], color="r", align="center")
plt.xticks(range(len(indices)), feat_names[indices], rotation='vertical')
plt.xlim([-1, len(indices)])
plt.show()
```
